```python
# method to print Json Object in readable format.. 
def printJsonObject(obj):
    line_list = obj.split("\n")
    for line in line_list:
        print(line)
```

```python
# method to get category
def getCategoryFromJson(data):
    category = {}
    for i, row in data.iterrows():
        tempdict = dict(row)
        category[tempdict["items"]["id"]]= tempdict["items"]["snippet"]["title"]
    return category
```

```python
# method to get combine dataFrames 
def appendCategoryToDataFrame(dataFrame,category):
    tempList = []
    for i,row in dataFrame.iterrows():
        if str(row["category_id"]) in category.keys():
            tempList.append(category[str(row["category_id"])])
        else :
            tempList.append(None)
    dataFrame["category_name"] = tempList
    return dataFrame
```

### Most Common Tags used across Trending Videos!!

```python
## Word Frequency of tags and sorted in decending order
tagsDictionary = {}
for i,row in videoCompleteDF.iterrows():
    for tag in row["tags"].split('|'):
        if tag in tagsDictionary:
            tagsDictionary[tag] += 1
        else:
            tagsDictionary[tag] = 1
sortedDict = dict(sorted(tagsDictionary.items(), key=lambda x:x[1], reverse=True))
sortedDict
```

```python
## Plot word frequencey graph for top 50 most occuring word
wordsList = []
countList = []
for key,val in sortedDict.items():
    wordsList.append(key)
    countList.append(val)
fig, axs=plt.subplots(figsize=(20,10))
axs.plot(wordsList[:50],countList[:50])
axs.set_xticklabels(wordsList[:50],fontsize=10)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
## Word Cloud of most frequently used Tags
cloud = WordCloud(max_font_size=80,colormap="hsv").generate_from_frequencies(tagsDictionary)
plt.figure(figsize=(16,12))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
```

<h2>Preparing data for deeper analysis</h2>


```python


grouped_data = Main_videos.groupby(['comments_disabled', 'ratings_disabled']).agg(
    average_views=('views', 'mean'),
    total_videos=('video_id', 'count')
).reset_index()

grouped_data
```

<h3>videoDF.groupby(['comments_disabled', 'ratings_disabled']):</h3>
This part of the code groups the DataFrame by the values in the comments_disabled and ratings_disabled columns.
groupby creates a grouping of rows that have the same values in the specified columns. Here, it's creating groups based on the combination of whether comments and ratings are disabled or enabled for each video.

<h3>.agg(...).</h3>
It's used here to apply one or more operations to each group of data.
Inside the agg method, a dictionary is passed which specifies the operations to be applied to specific columns.


<h3>average_views=('views', 'mean'):</h3>
This part of the dictionary is specifying that the mean (average) of the 'views' column should be calculated for each group.
The result of this calculation is labeled as average_views in the resulting aggregated DataFrame.

<h3>total_videos=('video_id', 'count'): </h3>
Similarly, this part is specifying that the count of 'video_id' should be calculated for each group, which gives the number of videos in each group.
This result is labeled as total_videos in the aggregated DataFrame.

<h3>.reset_index():</h3>
 After the aggregation, reset_index is used to reset the index of the DataFrame.
This step converts the multi-level index created by the groupby into regular columns, making the DataFrame easier to work with.


<h2>Plotting the results</h2>

```python
# # Plotting the results
plt.figure(figsize=(10, 6))
grouped_data['comments_disabled'] = grouped_data['comments_disabled'].astype(str)
grouped_data['ratings_disabled'] = grouped_data['ratings_disabled'].astype(str)

# Creating a bar plot for average views
sns.barplot(x='comments_disabled', y='average_views', hue='ratings_disabled', data=grouped_data)

plt.title('Average Views by Comments and Ratings Status')
plt.xlabel('Comments Disabled')
plt.ylabel('Average Views')
plt.xticks([0, 1], ['Enabled', 'Disabled'])
plt.legend(title='Ratings Disabled', loc='upper right')

plt.show()

# Plotting the number of videos in each category
plt.figure(figsize=(10, 6))

sns.barplot(x='comments_disabled', y='total_videos', hue='ratings_disabled', data=grouped_data)

plt.title('Number of Videos by Comments and Ratings Status')
plt.xlabel('Comments Disabled')
plt.ylabel('Number of Videos')
plt.xticks([0, 1], ['Enabled', 'Disabled'])
plt.legend(title='Ratings Disabled', loc='upper right')

plt.show()
```
